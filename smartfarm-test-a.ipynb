```python
import pandas as pd

file_path = r"C:\Users\rpaeh\Desktop\port_information_20231212.csv"

data = pd.read_csv(file_path, encoding='euc-kr')

print(data.head())
```

```python
port_info = ['항만명', '항만위치', '항구역', '수(물)면적', '부지면적']
port_data = data[port_info]

print(port_data.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
keep_columns = ['수(물)면적', '부지면적']
filtered_df = data[keep_columns]

correlation_matrix = filtered_df.corr()

cmap = ListedColormap(['#A0D3E8', '#D3D3D3'])

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap=cmap, fmt='.2f', cbar=False)
plt.title('수(물)면적과 부지면적 상관계수 히트맵')
plt.show()
```

```python
latitude = 35.1796  # 부산 위도
longitude = 129.0756  # 부산 경도

# 지도 생성
map_busan = folium.Map(location=[latitude, longitude], zoom_start=10)

# 항만 정보 추가
for index, row in data.iterrows():
    folium.Marker(
        location=[latitude, longitude],
        popup=f"{row['항만명']}: {row['주요취급화물']}",
        icon=folium.Icon(color='blue')
    ).add_to(map_busan)

# 지도 출력
map_busan
```
